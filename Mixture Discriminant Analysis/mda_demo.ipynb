# Mixture Discriminant Analysis

### Synthetic Example

```python
    import os
    os.chdir('/Users/amazaspshaumyan/Desktop/MixtureExperts/Mixture Discriminant Analysis')
    import numpy as np
    import matplotlib.pyplot as plt
    from sklearn.cross_validation import train_test_split
    from mda import MDA
    from sklearn.lda import LDA
    from sklearn.qda import QDA
    from sklearn.metrics import confusion_matrix
    %matplotlib inline
    
    # generate data set
    np.random.seed(0)
    X_train   = np.ones([600,2])
    X_test    = np.ones([600,2])
    Y_train   = np.zeros(600)
    Y_test    = np.zeros(600)
    centroids = [[0,0],[0,20],[10,10],[10,0],[0,10],[10,20]]
    m         = 100
    for i,mu in enumerate(centroids):
        sample   = 2*np.random.randn(2*m,2) + np.array(mu)
        X_train[i*m:(i+1)*m,:],X_test[i*m:(i+1)*m,:] = train_test_split(sample,test_size = 0.5)
    Y_train[300:600] = 1
    Y_test[300:600]  = 1
        
    
    # plot training data 
    plt.plot(X_train[Y_train==1,0],X_train[Y_train==1,1],"bo", markersize = 5, label = "1")
    plt.plot(X_train[Y_train==0,0],X_train[Y_train==0,1],"ro", markersize = 5, label = "0")
    plt.legend()
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("Training set plot")
    plt.show()
    
    # fit mda
    mda   = MDA(Y_train,X_train,k = 2,clusters = [3,3], verbose = False)
    mda.fit()
    y_mda = mda.predict(X_test)
    
    # fit lda
    lda   = LDA()
    lda.fit(X_train,Y_train)
    y_lda = lda.predict(X_test)

    
    # fit qda
    qda   = QDA()
    qda.fit(X_train,Y_train)
    y_qda = qda.predict(X_test)

    # construct grid 
    x1         = np.linspace(-7,17,500)
    x2         = np.linspace(-7,28,500)
    X1,X2      = np.meshgrid(x1,x2)
    Xgrid      = np.zeros([250000,2])
    Xgrid[:,0] = np.reshape(X1,(250000,))
    Xgrid[:,1] = np.reshape(X2,(250000,))
    
    # predict on grid data
    prob_mda   = mda.predict_probs(Xgrid)
    prob_lda   = lda.predict_proba(Xgrid)
    prob_qda   = qda.predict_proba(Xgrid)
    
    models = [prob_mda,prob_lda,prob_qda]
    titles = ["MDA","LDA","QDA"]
    for i,model in enumerate(models):
      plt.contourf(X1,X2,np.reshape(model[:,0],(500,500)), cmap="coolwarm")
      plt.colorbar()
      plt.plot(X_test[Y_test==0,0],X_test[Y_test==0,1],"ro", markersize = 3)
      plt.plot(X_test[Y_test==1,0],X_test[Y_test==1,1],"bo", markersize = 3)
      title = titles[i]+" decision boundaries on test set"
      plt.title(title)
      plt.xlabel("x1")
      plt.ylabel("x2")
      plt.show()
        
       
    # confusion matrices for all three methods above
    print "RESULTS ON TEST SET \n"
    print "\n LDA \n"
    print confusion_matrix(Y_test,y_lda)
    print "\n QDA \n"
    print confusion_matrix(Y_test,y_qda)
    print "\n MDA \n"
    print confusion_matrix(Y_test,y_mda)
    print "\n"
    
```

### Example with Iris data

```python

# load data
np.random.seed(0)
from sklearn import datasets
iris = datasets.load_iris()
iris_train,iris_test,y_train,y_test = train_test_split(iris.data,iris.target, test_size = 0.25)

# mda
mda_iris      = MDA(y_train,iris_train, k = 3, clusters = [2,2,2], verbose = False)
mda_iris.fit()
y_test_mda    = mda_iris.predict(iris_test)
print "MDA \n"
print confusion_matrix(y_test_mda,y_test)

# lda
lda           = LDA()
lda.fit(iris_train,y_train)
y_test_lda    = lda.predict(iris_test)
print "LDA \n"
print confusion_matrix(y_test_lda,y_test)

# qda 
qda           = QDA()
qda.fit(iris_train,y_train)
y_test_qda    = qda.predict(iris_test)
print "QDA \n"
print confusion_matrix(y_test_qda,y_test)
```
