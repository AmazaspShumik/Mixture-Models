# Clustering Examples

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from vbgmmard import VBGMMARD
%matplotlib inline 
```

## Old Faithful Geyser Example

```python
Data = pd.read_csv('old_faithful.txt')
gmm  = VBGMMARD(max_components = 20, init_type = 'auto')
gmm.fit(np.array(Data))
```

```python
plt.figure(figsize = (14,8))
plt.plot(Data['eruptions'],Data['waiting'],'b+',markersize = 7, label = 'data')
plt.plot(gmm.means[:,0], gmm.means[:,1],'ro',markersize = 12, label = 'final centoids')
plt.plot(gmm.means0[:,0], gmm.means0[:,1],'go', label = 'initial centroids')
plt.title("Old Faithful clustering")
plt.xlabel('eruptions')
plt.ylabel('waiting')
plt.legend(loc = 2)
plt.show()
```
