# Clustering with Bayesian GMM with Automatic Model Selection

In this notebook we demonstrate use of Variational Bayesian GMM with automatic model selection. Model that we use is not fully Bayesian, weights are considered to be parameters, so we iterativel use Variational Bayes to approximate Lower Bound and then maximize Lower Bound with respect to weight parameter. In the process of iterations some weights are driven to zero and are pruned if they achieve threshold level (by default 1e-5).

References:
==========
Adrian Corduneanu and Chris Bishop, Variational Bayesian Model Selection for Mixture Distributions (2001)

Theoretical Note:
=================
In original paper only one iteration of mean field approximation is used before maximizing lower bound with respect to
weighting parameters, in these examples we use 10 by default. With one iteration algorithm still gives good results but they seem to be quite unstable (most likely because of bad approximation), but with better approximation which is achieved using 30 iterations results seems to be much more stable (although obviously at the cost of time)

```python
import numpy as np
import matplotlib.pyplot as plt
from vbgmmard import VBGMMARD
from matplotlib.patches import Ellipse
import pandas as pd
%matplotlib inline



def plotter(X, max_k,title, rand_state = 1, prune_thresh = 1e-5, mfa_max_iter = 10):
    '''
    Plotting function for VBGMMARD clustering
    
    Parameters:
    -----------
    X: numpy array of size [n_samples, n_features]
       Data matrix
       
    max_k: int
       Maximum number of components
       
    title: str
       Title of the plot
       
    Returns:
    --------
    :instance of VBGMMARD class 
    '''
    # fit model & get parameters
    gmm = VBGMMARD(max_components = max_k, prune_thresh = prune_thresh,
                   mfa_max_iter = mfa_max_iter, rand_state = rand_state)
    gmm.fit(X)
    params  = gmm.get_params()
    centers = params['means']
    covars  = params['covars']
    k_selected = centers.shape[0]
    
    # plot data
    fig, ax = plt.subplots(figsize = (10,6))
    ax.plot(X[:,0],X[:,1],'bo', label = 'data')
    ax.plot(centers[:,0],centers[:,1],'rD', markersize = 8, label = 'means')
    for i in range(k_selected):
        plot_cov_ellipse(pos = centers[i,:], cov = covars[i], ax = ax)
    
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend(loc = 2)
    plt.title((title + ', {0} initial clusters, {1} selected clusters').format(max_k,k_selected))
    plt.show()
    return gmm
    
    
# plot_cov_ellipse function is taken from  
# https://github.com/joferkington/oost_paper_code/blob/master/error_ellipse.py

def plot_cov_ellipse(cov, pos, nstd=2, ax=None, **kwargs):
    """
    Plots an `nstd` sigma error ellipse based on the specified covariance
    matrix (`cov`). Additional keyword arguments are passed on to the 
    ellipse patch artist.

    Parameters
    ----------
        cov : The 2x2 covariance matrix to base the ellipse on
        pos : The location of the center of the ellipse. Expects a 2-element
            sequence of [x0, y0].
        nstd : The radius of the ellipse in numbers of standard deviations.
            Defaults to 2 standard deviations.
        ax : The axis that the ellipse will be plotted on. Defaults to the 
            current axis.
        Additional keyword arguments are pass on to the ellipse patch.

    Returns
    -------
        A matplotlib ellipse artist
    """
    def eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]

    if ax is None:
        ax = plt.gca()

    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))

    # Width and height are "full" widths, not radius
    width, height = 2 * nstd * np.sqrt(vals)
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, fill = False,
                    edgecolor = 'k',linewidth = 4,**kwargs)

    ax.add_artist(ellip)
    return ellip
```

## Old Faithful Geyser data

```python
Data = pd.read_csv("old_faithful.txt")
Data = np.array(Data[['eruptions','waiting']])
geyser_model_gmm1  = plotter(Data, 20,'Old Faithful')
```

```python
geyser_model_gmm2  = plotter(Data, 20,'Old Faithful', mfa_max_iter = 1)
```

## Synthetic Data Example 1

```python
X = np.zeros([600,2])
np.random.seed(0)
X[0:200,:]   = np.random.multivariate_normal(mean = (0,10), cov = [[3,0],[0,2]], size = 200)
X[200:400,:] = np.random.multivariate_normal(mean = (0,0) , cov = [[3,0],[0,2]], size = 200)
X[400:600,:] = np.random.multivariate_normal(mean = (0,-10) , cov = [[3,0],[0,2]], size = 200)

sy_gmm_1 = plotter(X,20,'Synthetic Example')
```

## Synthetic Data Example 2 

```python
X = np.zeros([600,2])
np.random.seed(0)
X[0:200,:]   = np.random.multivariate_normal(mean = (0,0), cov = [[10,0],[0,1]], size = 200)
X[200:400,:] = np.random.multivariate_normal(mean = (0,0), cov = [[1,0],[0,10]], size = 200)
X[400:600,:] = np.random.multivariate_normal(mean = (0,0), cov = [[10,-10],[10,-10]], size = 200)
sy_gmm_2 = plotter(X,20,'Synthetic Example', prune_thresh = 1e-5, mfa_max_iter = 10)
```
