# Examples of applying HME to multimodal output

### Example No 1

```python
    import matplotlib.pyplot as plt
    import numpy as np
    import general_hme as hm
    % matplotlib inline
    
    X            = np.zeros([600,1])
    Y            = np.zeros(600)
    X[0:400,0]   = np.linspace(-3,3,400)
    X[400:600,0] = np.linspace(-3,3,200)
    Y[0:400]     = -1*X[0:400,0]**2 + 5 + np.random.normal(0,1,400)
    Y[400:600]   = X[400:600,0]**2 - 5 + np.random.normal(0,1,200)
    
    # plot data
    plt.plot(X[:,0],Y,"c+")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
hme = hm.HME(Y, X,Y,X,"gaussian",bias = True, gate_type = "softmax", levels = 2, branching = 7)
hme.fit()
```

```python
    # calculate probability of observing points in squares formed by grid 
    y_lo = -10*np.ones(400)
    y_hi = 10*np.ones(400)
    x1,x2,p = hm.prob_grid(hme,X[0:400,:],y_lo,y_hi,n_steps = 200,posterior_type = "pdf")
    n,m = np.shape(x1)
    
    # plot raw probabilities, pdf
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.title("Probabilities, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
    
    # plot of original data along with pdf
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.plot(X[:,0],Y,"c+")
    plt.title("Probabilities & Original Data, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
    # CDF
    y_lo_two = -10*np.ones(400)
    y_hi_two = 10*np.ones(400)
    X1,X2,P = hm.prob_grid(hme,X[0:400,:],y_lo_two,y_hi_two,n_steps = 200, posterior_type = "cdf")
    
    # plot raw probabilities, cdf
    plt.pcolor(X1,X2,P,cmap = "coolwarm")
    plt.title("Probabilities, cdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

### Example No 2: Sinusoids

```python
    X = np.zeros([1000,1])
    Y = np.zeros(1000)
    thr0 = 500; thr1 = 1000
    X[0:thr0,0]    = np.linspace(0,6,thr0)
    X[thr0:thr1,0] = np.linspace(0,6,thr0)
    Y[0:thr0] = 6*np.sin(X[0:thr0,0]) + np.random.normal(0,1,thr0)
    Y[thr0:thr1] = -6*np.sin(X[thr0:thr1,0]) + np.random.normal(0,1,thr0)
```

```python
plt.plot(X[:,0],Y,"c+")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original data')
```

```python
    hme = hm.HME(Y,X,Y,X,"gaussian",bias = True, gate_type = "wgda",levels = 4,branching = 2, verbose = False)   
    hme.fit()
```

```python
    y_lo = -10*np.ones(thr0)
    y_hi = 10*np.ones(thr0)
    x1,x2,p = hm.prob_grid(hme,X[0:thr0,:],y_lo,y_hi,n_steps = 200,posterior_type = "pdf")    
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.title("Probabilities, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.plot(X[:,0],Y,"c+")
    plt.title("Probabilities & Original Data, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
    y_lo_two = -8*np.ones(thr0)
    y_hi_two = 10*np.ones(thr0)
    X1,X2,P = hm.prob_grid(hme,X[0:thr0,:],y_lo_two,y_hi_two,n_steps = 200, posterior_type = "cdf")
    plt.pcolor(X1,X2,P,cmap = "coolwarm")
    plt.title("Probabilities, cdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```
