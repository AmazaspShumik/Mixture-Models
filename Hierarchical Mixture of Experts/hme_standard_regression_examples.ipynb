# Examples of applying HME to Regression Problems 

### Example 1:  sinc(x)

```python
    import os
    os.chdir('/Users/amazaspshaumyan/Desktop/MixtureExperts/Hierarchical Mixture of Experts')
    
    import numpy as np
    import matplotlib.pyplot as plt
    import general_hme as hm
    from sklearn.cross_validation import train_test_split
    from sklearn.metrics import mean_squared_error
    from sklearn.svm import SVR
    from sklearn.ensemble import GradientBoostingRegressor
    from sklearn.grid_search import GridSearchCV
    %matplotlib inline
    
    np.random.seed(0)
    # Simulate scaled sinc data
    X           =  np.zeros([2000,1])
    X[:,0]      = np.linspace(-5, 5, 2000)
    Y           = 15*np.sinc(X[:,0]) + np.random.normal(0,1,2000)
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.4, 
                                                          random_state = 0)

    # fit hme model
    hme     = hm.HME(Y_train, X_train,Y_test,X_test,"gaussian",bias = True, 
                                                                    gate_type = "wgda",
                                                                    verbose = False, 
                                                                    levels = 8, 
                                                                    branching  = 2)
    hme.fit()
    
    # fit SVR 
    svr     = GridSearchCV(SVR(kernel='rbf'),param_grid={"gamma":np.logspace(-2,2,5),
                                                      "C":[0.01,1,10,100,1000]}, 
                                                      cv=5)
    svr.fit(X_train,Y_train)
    
    # fit Gradient Boosting Regression
    gb      = GradientBoostingRegressor()
    gbr     = GridSearchCV(gb , param_grid = {"learning_rate":[0.1,1,5,10],
                                              "max_depth":[3,5,7]},
                                              cv = 5)
    gbr.fit(X_train,Y_train)
    
    # predict with different models
    gbr_predict  = gbr.predict(X_test)
    svr_predict  = svr.predict(X_test)
    hme_predict  = hme.predict(X_test)
    
    mse = [mean_squared_error(e,Y_test) for e in [hme_predict,svr_predict,gbr_predict]]
    
    print "TEST SET RESULTS \n"
    print "MSE for Hierarchical Mixture Models is {0} \n".format(mse[0])
    print "MSE for Support Vector Regression is {0} \n".format(mse[1])
    print "MSE for Gradient Boosting Regression is {0} \n".format(mse[2])
    
    # plot point predictor
    plt.figure(figsize = (7,7))
    plt.plot(X_test[:,0],Y_test,"bo", label = "test data")
    plt.plot(X_test[:,0],hme_predict,"rD", label = "prediction")
    plt.title("Point predictor by hierarchical mixture model")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc = 1)
    plt.show()
```

### Boston Housing Example

```python
from sklearn import datasets


np.random.seed(1)
boston                                  = datasets.load_boston()
boston_train,boston_test,y_train,y_test = train_test_split(boston.data,boston.target,
                                                                       test_size = 0.2,
                                                                       random_state = 1)

# Gradient Boosting Regression
gb                 = GradientBoostingRegressor()
gbr                = GridSearchCV(gb,param_grid = { "learning_rate":[0.01,0.1,1],
                                                    "max_depth":[1,5,10],
                                                    "n_estimators":[100,200,400]},
                                                    cv = 5)
gbr.fit(boston_train,y_train)
gbr_pred_b         = gbr.predict(boston_test)


# Support Vector Regression
svr                = GridSearchCV(SVR(kernel='rbf'),param_grid={"gamma":np.logspace(-2,2,5),
                                                                "C":[0.01,1,10,100,1000]}, 
                                                                cv=5)
svr.fit(boston_train,y_train)
svr_pred_b         = svr.predict(boston_test)


# Hierarchical Mixture of Experts
hme                = hm.HME(y_train, boston_train,y_test,boston_test,"gaussian",
                                                                     bias = True, 
                                                                     gate_type = "softmax",
                                                                     verbose = False, 
                                                                     levels = 2,
                                                                     branching  = 3)
hme.fit()
hme_pred_b        = hme.predict(boston_test)

print "BOSTON HOUSING TEST DATA RESULTS \n"
print "MSE on test data for HME is {0} \n".format(mean_squared_error(hme_pred_b,y_test))
print "MSE on test data for GBR is {0} \n".format(mean_squared_error(gbr_pred_b,y_test))
print "MSE on test data for SVR is {0} \n".format(mean_squared_error(svr_pred_b,y_test))                                        


```

### HME drawback, non-convex optimization

```python
    X      = np.ones([1500,1])
    X[:,0] = np.linspace(0, 10, 1500)
    Y = np.zeros(1500)
    Y[0:500]  = X[0:500,0]*4 + np.random.normal(0,1,500) + 100*np.ones(500)
    Y[500:1000] = X[500:1000,0]*(-4) + np.random.normal(0,1,500) + 200*np.ones(500)
    Y[1000:1500] = X[1000:1500,0]*8 + np.random.normal(0,1,500) + 160*np.ones(500)
    hme = hm.HME(Y, X,Y,X,"gaussian",bias = True, gate_type = "softmax",verbose = False,
                                                                        levels = 4, 
                                                                        branching = 4)
    hme.fit()
    Y_hat = hme.predict(X)
    plt.plot(X[:,0],Y,"bo", label = "data")
    plt.plot(X[:,0],Y_hat,"r-", label = "fitted model", linewidth = 3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc = 2)
    plt.title("Local Maximum")
    plt.show()
```
