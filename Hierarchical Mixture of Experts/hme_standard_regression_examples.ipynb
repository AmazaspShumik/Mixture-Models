# Examples of applying HME to Regression Problems 

### Example N 1

```python
    import numpy as np
    import matplotlib.pyplot as plt
    import general_hme as hm
    %matplotlib inline
    
    
    # Simulate sinusoid data set
    X_train   =  np.zeros([3000,1])
    X_test    =  np.zeros([1000,1])
    
    # training & test set
    X_train[:,0] = np.linspace(0, 10, 3000)
    X_test[:,0]  = np.random.random(1000)*10
    Y_train      = np.sin(X_train[:,0])*4 + np.random.normal(0,1,3000)
    Y_test       = np.sin(X_test[:,0])*4 + np.random.normal(0,1,1000)

    # fit hme model
    hme = hm.HME(Y_train, X_train,Y_test,X_test,"gaussian",bias = True, gate_type = "softmax",verbose = False, levels = 2, branching  = 4)
    hme.fit()
    
    # plot point predictor
    Y_hat = hme.predict(X_train)
    plt.plot(X_train[:,0],Y_train,"bo", label = "data")
    plt.plot(X_train[:,0],Y_hat,"r+", linewidth = 3, label = 'fitted model')
    plt.title("Point predictor by hierarchical model")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc = 4)
    plt.show()
    
    # craete grid 
    y_lo = -6*np.ones(3000)
    y_hi = 6*np.ones(3000)
    x1,x2,p = hm.prob_grid(hme,X_train,y_lo,y_hi,n_steps = 400,posterior_type = "pdf")
    
    # plot probability distribution over grid
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.title("Probabilities, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
    deep_hme = hm.HME(Y_train, X_train,Y_test,X_test,"gaussian",bias = True, gate_type = "wgda",verbose = False, levels = 8, branching = 2)   
    deep_hme.fit()

    # craete grid 
    y_lo = -6*np.ones(3000)
    y_hi = 6*np.ones(3000)
    x1,x2,p = hm.prob_grid(deep_hme,X_train,y_lo,y_hi,n_steps = 300,posterior_type = "pdf")
    
    # plot probability distribution over grid
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.title("Probabilities, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

### Example N 2

```python
    X = np.zeros([3000,1])
    X[:,0] = np.linspace(0,12.3,3000)
    Y = (X[:,0]-1)*X[:,0]*(X[:,0]-2)*(X[:,0]-8)*(X[:,0]-9.8)*(X[:,0]-12) + np.random.normal(0,400,3000)
    hme = hm.HME(Y, X,Y[0:10],X[0:10,:],"gaussian",bias = True, gate_type = "wgda",verbose = False, levels = 8, branching = 2)   
    hme.fit()
    
    Y_hat = hme.predict(X)
    plt.plot(X[:,0],Y,"bo", label = "data")
    plt.plot(X[:,0],Y_hat,"r-", linewidth = 3, label = "fitted model")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc = 2)
    plt.show()
```

### Example N 3

```python
    X      = np.ones([1500,1])
    X[:,0] = np.linspace(0, 10, 1500)
    Y = np.zeros(1500)
    Y[0:500]  = X[0:500,0]*4 + np.random.normal(0,1,500) + 100*np.ones(500)
    Y[500:1000] = X[500:1000,0]*(-4) + np.random.normal(0,1,500) + 200*np.ones(500)
    Y[1000:1500] = X[1000:1500,0]*8 + np.random.normal(0,1,500) + 160*np.ones(500)
    hme = hm.HME(Y, X,Y[0:10],X[0:10,:],"gaussian",bias = True, gate_type = "softmax",verbose = False, levels = 3, branching = 4)
    hme.fit()
    Y_hat = hme.predict(X)
    plt.plot(X[:,0],Y,"bo", label = "data")
    plt.plot(X[:,0],Y_hat,"r-", label = "fitted model", linewidth = 3)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc = 2)
    plt.show()
```
