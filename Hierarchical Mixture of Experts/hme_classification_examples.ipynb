# Examples of applying HME to Classification Problems

### Example with synthetic data

```python
    import numpy as np
    import matplotlib.pyplot as plt
    import general_hme as hm
    from sklearn.cross_validation import train_test_split
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.grid_search import GridSearchCV
    from sklearn.metrics import confusion_matrix
    from sklearn.svm import SVC
    %matplotlib inline

    # create data set
    np.random.seed(0)
    X     = np.random.random([4000,2])
    Y     = np.array(["y" for i in range(4000)])
    Y[(X[:,0]-0.5)**2+(X[:,1]-0.5)**2 < 0.1] = "n"
    X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1, random_state = 0) 
    
    # fit hme model
    hme   = hm.HME(Y_train,X_train,Y_test,X_test,"wgda",bias = True, 
                                                        gate_type = "wgda",
                                                        verbose = False, 
                                                        levels = 6,
                                                        branching = 2)
    hme.fit()
    hme_predict = hme.predict(X_test)
    
    # fit svm model
    svc   = GridSearchCV(SVC(kernel = "rbf"), param_grid = {"C":[0.01,0.1,1,10],
                                                            "gamma":[0.01,0.5,1,2,5,10,100]},
                                                           cv = 5)
    svc.fit(X_train,Y_train)
    svc_predict = svc.predict(X_test)
    
    
    # fit random forest classifier
    rfc   = GridSearchCV(RandomForestClassifier(), param_grid = {"max_features":[1,2]},
                                                                  cv = 5)
    rfc.fit(X_train,Y_train)
    rfc_predict = rfc.predict(X_test)
    
    print "RESULTS ON TEST DATA \n"
    print "SVM \n"
    print confusion_matrix(svc_predict,Y_test)
    print "\n Random Forest \n"
    print confusion_matrix(rfc_predict,Y_test)
    print "\n HME \n"
    print confusion_matrix(hme_predict,Y_test)
    
    
    # grid
    x1         = np.linspace(0,1,500)
    x2         = np.linspace(0,1,500)
    X1,X2      = np.meshgrid(x1,x2)
    Xgrid      = np.zeros([250000,2])
    Xgrid[:,0] = np.reshape(X1,(250000,))
    Xgrid[:,1] = np.reshape(X2,(250000,))
    
    # predict hme probs on grid
    hme_grid_predict = hme.predict(Xgrid, predict_type = "predict_probs")
    
    # plot grid
    plt.figure(figsize = (7,7))
    plt.contourf(X1,X2,np.reshape(hme_grid_predict[:,0],(500,500)), cmap="coolwarm")
    plt.plot(X_test[Y_test=="y",0],X_test[Y_test=="y",1],"bo", markersize = 3)
    plt.plot(X_test[Y_test=="n",0],X_test[Y_test=="n",1],"ro", markersize = 3)
    plt.colorbar()
    title = "HME decision boundaries on test set"
    plt.title("Probabilities HME")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.show()
    
    
    
```

### Another example with synthetic data

```python
    
    X = np.random.random([4000,2])
    Y = np.array(["a" for i in range(4000)])
    Y[(X[:,0])**2+(X[:,1])**2 < 0.09] = "b"
    Y[(X[:,0]-1)**2+(X[:,1]-1)**2 < 0.09] = "c"
    Y[(X[:,0])**2+(X[:,1]-1)**2 < 0.09] = "d"
    Y[(X[:,0]-1)**2+(X[:,1]-1)**2 < 0.09] = "e"
    Y[(X[:,0]-1)**2+(X[:,1])**2 < 0.09] = "f"

    
    hme = hm.HME(Y, X,Y_test,X_test,"wgda",bias = True, gate_type = "wgda",verbose = False, levels = 9, branching = 2)
    hme.fit()
    Y_hat = hme.predict(X, predict_type = "predict_response")
    color_type = ["bo","ro","go","co","yo","mo"]
    
    plt.figure(figsize = (10,10))
    for index,class_id in enumerate(["a","b","c","d","e","f"]):
        plt.plot(X[Y_hat==class_id,0],X[Y_hat==class_id,1],color_type[index],label = class_id)
    plt.plot(X[Y_hat!=Y,0],X[Y_hat!=Y,1],"ko",label = "misclassification", markersize = 10)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc=2,prop={'size':9})
    plt.title("Classification example")
    plt.show()
```
