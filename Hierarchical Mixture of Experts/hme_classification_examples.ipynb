# Examples of applying HME to Classification Problems

### Example N 1

```python
    import numpy as np
    import matplotlib.pyplot as plt
    import general_hme as hm
    from sklearn.cross_validation import train_test_split
    %matplotlib inline

    # create data set
    X_all = np.random.random([4000,2])
    Y_all = np.array(["y" for i in range(4000)])
    Y_all[(X_all[:,0]-0.5)**2+(X_all[:,1]-0.5)**2 < 0.1] = "n"
    X,X_test,Y,Y_test = train_test_split(X_all,Y_all,test_size = 0.6) 
    
    # fit model with weighted gaussian discriminant analysis
    hme_wgda = hm.HME(Y, X,Y_test,X_test,"wgda",bias = False, gate_type = "wgda",verbose = False, levels = 6, branching = 2)
    hme_wgda.fit()
    
    # fit model with softmax regression
    hme_softmax = hm.HME(Y, X,Y_test,X_test,"softmax",bias = True, gate_type = "softmax",verbose = False, levels = 6, branching = 2)
    hme_softmax.fit()
    
    # plot model that used wgda ( training set)
    Y_hat_gd = hme_wgda.predict(X, predict_type= "predict_response")
    plt.plot(X[Y_hat_gd=="n",0],X[Y_hat_gd=="n",1],"ro",label = "y")
    plt.plot(X[Y_hat_gd=="y",0],X[Y_hat_gd=="y",1],"bo",label = "n")
    plt.plot(X[Y_hat_gd!=Y,0],X[Y_hat_gd!=Y,1],"go",label = "misclassified")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.title("Classification with WGDA")
    plt.show()
    
    # plot model that used softmax (training set)
    Y_hat_sr = hme_softmax.predict(X, predict_type= "predict_response")
    plt.plot(X[Y_hat_sr=="n",0],X[Y_hat_sr=="n",1],"ro",label = "y")
    plt.plot(X[Y_hat_sr=="y",0],X[Y_hat_sr=="y",1],"bo",label = "n")
    plt.plot(X[Y_hat_sr!=Y,0],X[Y_hat_sr!=Y,1],"go",label = "misclassified")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.legend()
    plt.title("Classification with Softmax")
    plt.show()
    
```

```python
    # prediction on test set
    Y_hat_test_gd = hme_wgda.predict(X_test, predict_type= "predict_response")
    Y_hat_test_sr = hme_softmax.predict(X_test, predict_type = "predict_response")
    
    # construct confusion matrices
    from sklearn.metrics import confusion_matrix
    
    print "HME with softmax regression, training set"
    print confusion_matrix(Y,Y_hat_sr)
    print "HME with gaussian discriminant, training set"
    print confusion_matrix(Y,Y_hat_gd)
    
    print "HME with softmax regression, test set"
    print confusion_matrix(Y_test,Y_hat_test_sr)
    print "HME with gaussian discriminant, test set"
    print confusion_matrix(Y_test,Y_hat_test_gd)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

rf = RandomForestClassifier(n_estimators = 100)
rf.fit(X,Y)
Y_hat_test_rf = rf.predict(X_test)

gb = GradientBoostingClassifier( n_estimators = 100)
gb.fit(X,Y)
Y_hat_test_gb = gb.predict(X_test)

### make cross validation for random forest, gradient boosting
```

```python
print "Gradient Boosting Model"
print confusion_matrix(Y_test, Y_hat_test_gb)
print "Random Forest Classifier"
print confusion_matrix(Y_test, Y_hat_test_rf)
```

### Example N 2

```python
    
    X = np.random.random([2500,2])
    Y = np.array(["a" for i in range(2500)])
    Y[(X[:,0])**2+(X[:,1])**2 < 0.09] = "b"
    Y[(X[:,0]-1)**2+(X[:,1]-1)**2 < 0.09] = "c"
    Y[(X[:,0])**2+(X[:,1]-1)**2 < 0.09] = "d"
    Y[(X[:,0]-1)**2+(X[:,1]-1)**2 < 0.09] = "e"
    Y[(X[:,0]-0.5)**2+(X[:,1]-0.5)**2 < 0.09] = "f"

    
    hme = hm.HME(Y, X,Y_test,X_test,"wgda",bias = True, gate_type = "wgda",verbose = False, levels = 12, branching = 2)
    hme.fit()
    Y_hat = hme.predict(X, predict_type = "predict_response")
    color_type = ["bo","ro","go","co","yo","mo"]
    
    plt.figure(figsize = (10,10))
    for index,class_id in enumerate(["a","b","c","d","e","f"]):
        plt.plot(X[Y_hat==class_id,0],X[Y_hat==class_id,1],color_type[index],label = class_id)
    plt.plot(X[Y_hat!=Y,0],X[Y_hat!=Y,1],"ko",label = "misclassification")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend(loc=2,prop={'size':7})
    plt.title("Classification example")
    plt.show()
```
