# Examples of applying HME to multimodal output

### Example No1: Synthetic Data, Multimodal Output

```python
    import matplotlib.pyplot as plt
    import numpy as np
    import general_hme as hm
    from sklearn.cross_validation import train_test_split
    % matplotlib inline
    
    np.random.seed(0)
    X            = np.zeros([2000,1])
    Y            = np.zeros(2000)
    first_mode   = lambda x: -4*x**2 + 6 + 5*x
    second_mode  = lambda x: 4*x**2 - 6 + 5*x
    y            = np.zeros(10)
    r            = np.linspace(-2,2,200)
    probs        = []
    for i,el in enumerate(r):
        prob               = 0.3 + 0.4*float(i)/200
        probs.append(prob)
        lo                 = 10*i
        hi                 = 10*(i+1)
        I                  = np.random.random(10) < prob
        X[lo:hi,0]         = np.ones(10)*el
        for i,p in enumerate(I):
            if p:
                y[i]       = first_mode(X[lo+i,0])
            else:
                y[i]       = second_mode(X[lo+i,0])
        Y[lo:hi]           = y
    Y                      = Y + np.random.normal(0,1,2000)       
    
    
    # plot data
    plt.plot(X[:,0],Y,"c+")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Multimodal Output Data")
    plt.show()
    
    plt.plot(r,np.array(probs),"r-",linewidth = 3)
    plt.xlabel("x")
    plt.ylabel("Probability")
    plt.title("p(x)")
    plt.show()
    
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.5)
hme = hm.HME(Y_train, X_train,Y_test,X_test,"gaussian",bias = True, 
                                                       gate_type = "softmax",
                                                       verbose = False, 
                                                       levels = 8, 
                                                       branching  = 2)

hme.fit()

Xgrid       = np.ones([400,1])
Xgrid[:,0]  = np.linspace(-2,2,400)
y_lo        = -22*np.ones(400)
y_hi        = 22*np.ones(400)
x1,x2,p     = hm.prob_grid(hme,Xgrid,y_lo,y_hi,
                                          n_steps = 2000,
                                          posterior_type = "pdf")

# plot pdf heatmap
plt.figure(figsize=(7,7))
plt.pcolor(x1,x2,p,cmap = "coolwarm")
plt.title("Probabilities, pdf")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-22,22)


plt.figure(figsize=(7,7))
plt.pcolor(x1,x2,p,cmap = "coolwarm")
plt.title("Probabilities, pdf  & Test Data")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-22,22)
plt.plot(X_test[:,0],Y_test,"go", markersize = 3)
plt.show()
```

```python
Xgrid       = np.ones([400,1])
Xgrid[:,0]  = np.linspace(-2,2,400)
y_lo        = -22*np.ones(400)
y_hi        = 22*np.ones(400)
x1,x2,p     = hm.prob_grid(hme,Xgrid,y_lo,y_hi,
                                          n_steps = 2000,
                                          posterior_type = "cdf")

# plot pdf heatmap
plt.figure(figsize=(7,7))
plt.pcolor(x1,x2,p,cmap = "coolwarm")
plt.title("Probabilities y > Y > -22, cdf")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-22,22)


plt.figure(figsize=(7,7))
plt.pcolor(x1,x2,p,cmap = "coolwarm")
plt.title("Probabilities y > Y > -22, cdf  & Test Data")
plt.colorbar()
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-2,2)
plt.ylim(-22,22)
plt.plot(X_test[:,0],Y_test,"go", markersize = 3)
plt.show()
```

### Example No 2: Sinusoids

```python
    X = np.zeros([1000,1])
    Y = np.zeros(1000)
    thr0 = 500; thr1 = 1000
    X[0:thr0,0]    = np.linspace(0,6,thr0)
    X[thr0:thr1,0] = np.linspace(0,6,thr0)
    Y[0:thr0] = 6*np.sin(X[0:thr0,0]) + np.random.normal(0,1,thr0)
    Y[thr0:thr1] = -6*np.sin(X[thr0:thr1,0]) + np.random.normal(0,1,thr0)
```

```python
plt.plot(X[:,0],Y,"c+")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original data')
plt.show()
```

```python
    # fit data
    hme = hm.HME(Y,X,Y,X,"gaussian",bias = True, gate_type = "wgda",levels = 6,branching = 2, verbose = False)   
    hme.fit()
```

```python
    y_lo = -10*np.ones(thr0)
    y_hi = 10*np.ones(thr0)
    x1,x2,p = hm.prob_grid(hme,X[0:thr0,:],y_lo,y_hi,n_steps = 200,posterior_type = "pdf")    
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.title("Probabilities, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
    plt.pcolor(x1,x2,p,cmap = "coolwarm")
    plt.plot(X[:,0],Y,"co")
    plt.title("Probabilities & Original Data, pdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
    y_lo_two = -8*np.ones(thr0)
    y_hi_two = 10*np.ones(thr0)
    X1,X2,P = hm.prob_grid(hme,X[0:thr0,:],y_lo_two,y_hi_two,n_steps = 200, posterior_type = "cdf")
    plt.pcolor(X1,X2,P,cmap = "coolwarm")
    plt.plot(X[:,0],Y,"go", markersize = 3)
    plt.title("Probabilities y > Y > -8, cdf")
    plt.colorbar()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```
